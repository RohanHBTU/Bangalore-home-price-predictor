```python
df3['bhk']=df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3
```

```python
df3.drop('size',inplace=True,axis=1)
df3
```

```python
df3['total_sqft'].unique()
#we notice here that total area is not mentioned clearly and some ranges are also included
```

```python
#The try block lets you test a block of code for errors.
#The except block lets you handle the error.
#The else block lets you execute code when there is no error.
#The finally block lets you execute code, regardless of the result of the try- and except blocks.
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(convert_sqft_to_num)
```

```python
#we know that price per sqft is a very important feature in real estate industry, so let's create another feature 
df4['price_per_sqft']=df4['price']*100000/df4['total_sqft']
```

```python
df4
```

```python
df4['location']=df4['location'].apply(lambda x: x.strip()) #to strip out extra spaces 
location_stats=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
#we see here that there are too many localities having only one property so why not consider all of them as other
location_stats_less_than_10=location_stats[location_stats<=10]
```

```python
df4['location']=df4['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
df4[df4['total_sqft']/df4['bhk']<600].head()
```

```python
df4=df4[~(df4['total_sqft']/df4['bhk']<300)]
```

```python
df4.shape
```

```python
df4.price_per_sqft.describe()
```

```python
#we notice here that the maximum and minimum values for price per sqft are too much extreme which need to be handled.
def remove_pps_outliers(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df4=remove_pps_outliers(df4)
df4.shape
```
